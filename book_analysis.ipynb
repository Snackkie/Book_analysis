## Предсказывание средней цены книги

Для предсказания цены книги воспользуеся тремя регрессиями: Lasso, RandomForestRegressor, XGBRegressor. Сначала рассчитаем цену на начальных параметрах каждой модели, затем подберем подходящие. Ошибку будем мерять rmse.

```python
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from math import sqrt
```

```python
target_col = 'цена'
numerical_features = [x for x in df_reg.select_dtypes(include=np.number).columns.tolist()]
categorial_features = [x for x in df_reg.columns if x not in numerical_features and  x not in target_col]
```

```python
X = df_reg[numerical_features].drop(columns=['цена'])
Y = df_reg[target_col]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
est_lasso = Lasso(alpha= 10)
est_lasso.fit(X_train_scaled, Y_train)
sqrt(mean_squared_error(Y_test, est_lasso.predict(X_test_scaled)))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
lr_grid = {
    "alpha": np.logspace(-5, 3, 100)
}

lr_grid_searcher = GridSearchCV(Lasso(), lr_grid, cv=5, return_train_score=True, scoring="neg_mean_squared_error")
lr_grid_searcher.fit(X_train_scaled, Y_train)
sqrt(mean_squared_error(lr_grid_searcher.best_estimator_.predict(X_test_scaled), Y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor
est_rf = RandomForestRegressor(n_estimators= 100, max_depth= 70, max_features= 12)
est_rf.fit(X_train, Y_train)
sqrt(mean_squared_error(Y_test, est_rf.predict(X_test)))
```

```python
rf_grid = {
    "n_estimators": np.arange(100, 1001, 100),
    "max_depth": np.arange(50, 100, 5),
    "max_features": np.arange(5, 12, 4)
}

rf_grid_searcher = GridSearchCV(RandomForestRegressor(), rf_grid, cv=5, return_train_score=True, scoring="neg_mean_squared_error")
rf_grid_searcher.fit(X_train, Y_train)
sqrt(mean_squared_error(Y_test, rf_grid_searcher.best_estimator_.predict(X_test)))
```

```python
from xgboost import XGBRegressor
est_xgb = XGBRegressor(n_estimators=100, max_depth=7, eta=0.1, subsample=0.7, colsample_bytree=0.8)
est_xgb.fit(X_train, Y_train)
sqrt(mean_squared_error(Y_test, est_xgb.predict(X_test)))
```

```python
from hyperopt import STATUS_OK, Trials, fmin, hp, tpe
from hyperopt.pyll.base import scope
```

```python
XGBR_grid = {'max_depth': scope.int(hp.quniform("max_depth", 1, 5, 1)),
        'gamma': hp.uniform ('gamma', 0,1),
        'reg_alpha' : hp.uniform('reg_alpha', 0,50),
        'reg_lambda' : hp.uniform('reg_lambda', 10,100),
        'colsample_bytree' : hp.uniform('colsample_bytree', 0,1),
        'min_child_weight' : hp.uniform('min_child_weight', 0, 5),
        'n_estimators': scope.int(hp.quniform("n_estimators", 50, 1000, 1)),
        'learning_rate': hp.uniform('learning_rate', 0, .15),
        'tree_method':'gpu_hist',
        'gpu_id': 0,
        'random_state': 42,
        'max_bin' : scope.int(hp.quniform('max_bin', 200, 550, 1))}
```

```python
def hyperparameter_tuning_xgb_reg(space):
    model = XGBRegressor(**space)
    evaluation = [(X_train, Y_train), (X_test, Y_test)]
    model.fit(X_train, Y_train, eval_set=evaluation, eval_metric="rmse", early_stopping_rounds=100,verbose=False)
    pred = model.predict(X_test)
    rmse = sqrt(mean_squared_error(Y_test, pred))
    print ("SCORE:", rmse)
    return {'loss':rmse, 'status': STATUS_OK, 'model': model}
```

```python
trials = Trials()
best = fmin(fn=hyperparameter_tuning_xgb_reg,
            space=XGBR_grid,
            algo=tpe.suggest,
            max_evals=30,
            trials=trials)

print(best)
```

Наилучший результат показала модель XGBRegressor с ошибкой 302.8115437347564.

## Классификация

Изменим наш целевой признак для возможности его предсказывания классификаторами. Будем проверять 4 классификаторами: RandomForestClassifier, XGBClassifier, GaussianNB, VoteClassifier(из 3х предыдущих).

```python
target_col2 = df_main['Магазин'].apply(lambda x: 0 if x == 'Книжный лабиринт' else 1 if x =='Читай-город' else 2)
```

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
```

```python
X = df_reg[numerical_features].drop(columns=['Молодая гвардия', 'Читай-город', 'Книжный лабиринт'])
Y = target_col2
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)
X_test2, X_val, Y_test2, Y_val = train_test_split(X_test, Y_test, test_size=0.3)
```

```python
rf_clf1 = RandomForestClassifier()
rf_clf1.fit(X_train, Y_train)
accuracy_score(rf_clf1.predict(X_val), Y_val)
```

```python
RF_grid = {
    'max_depth': scope.int(hp.quniform("max_depth", 5, 40, 1)),
    'n_estimators': scope.int(hp.quniform("n_estimators", 10, 200, 1)),
    'max_features': scope.int(hp.quniform("max_features", 5, 10, 1)),
    'random_state': 42
}
```

```python
def hyperparameter_tuning_rf(space):
    model = RandomForestClassifier(**space)
    model.fit(X_train, Y_train)
    pred = model.predict(X_test2)
    acc = accuracy_score(Y_test2, pred)
    print ("SCORE:", 1 - acc)
    return {'loss':1 - acc, 'status': STATUS_OK, 'model': model}
```

```python
trials = Trials()
best = fmin(fn=hyperparameter_tuning_rf,
            space=RF_grid,
            algo=tpe.suggest,
            max_evals=30,
            trials=trials)
print(best)
```

```python
rf_clf = RandomForestClassifier(max_depth=9, max_features= 8, n_estimators= 124, random_state = 42)
rf_clf.fit(X_train, Y_train)
accuracy_score(rf_clf.predict(X_val), Y_val)
```

```python
xgb_clf = XGBClassifier()
evaluation = [(X_train, Y_train), (X_test2, Y_test2)]
xgb_clf.fit(X_train, Y_train, eval_set=evaluation, eval_metric="merror", early_stopping_rounds=100,verbose=False)
pred = xgb_clf.predict(X_val)
accuracy_score(Y_val, pred)
```

```python
XGB_clf_grid = {'max_depth': scope.int(hp.quniform("max_depth", 1, 10, 1)),
        'eta': hp.uniform ('eta', 0,1),
        'subsample' : hp.uniform('subsample', 0,1),
        'colsample_bytree' : hp.uniform('colsample_bytree', 0,1),
        'n_estimators': scope.int(hp.quniform("n_estimators", 50, 1000, 1)),
        'random_state': 42
        }
```

```python
def hyperparameter_tuning_xgb_clf(space):
    model = XGBClassifier(**space)
    evaluation = [(X_train, Y_train), (X_test2, Y_test2)]
    model.fit(X_train, Y_train, eval_set=evaluation, eval_metric="merror", early_stopping_rounds=100,verbose=False)
    pred = model.predict(X_test2)
    acc = accuracy_score(Y_test2, pred)
    print ("SCORE:", 1-acc)
    print ("params:", space)
    return {'loss':1-acc, 'status': STATUS_OK, 'model': model}
```

```python
trials = Trials()
best = fmin(fn=hyperparameter_tuning_xgb_clf,
            space=XGB_clf_grid,
            algo=tpe.suggest,
            max_evals=100,
            trials=trials)
print(best)
```

Функция, создающая возможный набор весов.

```python
a = np.linspace(0, 1, 11)
for i, d in enumerate(a):
    a1 = np.linspace(0, 1-d, 11 - i)
    a2 = np.linspace(1-d, 0, 11 - i)
    a3 = np.ones(11 - i)*d
    if i == 0:
        k = np.stack((a1, a2, a3), axis= -1)
    else:
        k = np.append(k, np.stack((a1, a2, a3), axis= -1), axis=0)
```

```python
clf_voting = VotingClassifier(estimators=[('rf', rf_clf), ('xgb', xgb_clf), ('bay', bay_clf)])
vote_grid = {
     "weights": k.tolist()
}
voting_grid_searcher = GridSearchCV(clf_voting, vote_grid, cv=3, return_train_score=True)
voting_grid_searcher.fit(X_train, Y_train)
```
